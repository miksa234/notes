```python
import numpy as np
```

```python
def lanczos_alg(A: np.ndarray, b: np.array, N: int=None) -> np.ndarray:
    if N == None:
        N = A.shape[0]
    v_1 = b/np.linalg.norm(b)
    V = np.zeros([A.shape[0], N])
    
    V[:,0] = v_1
    w̃_j = A @ V[:,0]
    α_j = np.conjugate(w̃_j) @ V[:,0]
    w_j = w̃_j - α_j*V[:,0]
    
    ritz = [np.conjugate(V[:,0]) @ A @ V[:,0]]
    for j in range(1, N):
        if (β_j:=np.linalg.norm(w_j)) != 0:
            V[:,j] = w_j/β_j
        else:
            break
        w̃_j = A @ V[:,j]
        α_j = np.conjugate(w̃_j) @ V[:,j]
        w_j = w̃_j - α_j*V[:,j] - β_j * V[:,j-1]
        
        ritz.append(np.conjugate(V[:,j]) @ A @ V[:,j])
    return V, ritz
```

```python
A = np.array([[4, 3, 2, 1],
              [3, 4, 3 ,2], 
              [2, 3, 4, 3], 
              [1, 2, 3, 4]])
b = np.random.rand(4)
V, ritz = lanczos_alg(A, b, N = 100)

print(V)
print('\n')
print(np.round(np.conjugate(V).T @ A @ V, 2))
print('\n')
print(np.round(ritz, 2)[-4:])
print(np.round(np.linalg.eig(A)[0], 2))
```
