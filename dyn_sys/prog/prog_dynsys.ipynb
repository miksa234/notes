```python
x = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, x)

# a)
u = 2*X + 3*Y
v = 1*Y

plt.quiver(X, Y, u, v)
```

```python
# b)
u = 2*X + 4*Y
v = -2*Y

plt.quiver(X, Y, u, v)
```

```python
# c)
X, Y, Z = np.meshgrid(x, x, x)
u = -1*X - 3*Y
v = 2*Y
w = -1*Z

ax = plt.figure().add_subplot(projection='3d')

ax.quiver(X, Y, Z, u, v, w, length=0.1, color = 'black')
```

```python
# d)
X, Y, Z = np.meshgrid(x, x, x)
u = -2*X - 3*Y
v = -1*Y
w = -1*Z

ax = plt.figure().add_subplot(projection='3d')

ax.quiver(X, Y, Z, u, v, w, length=0.1, color = 'black')
```

```python
#4c)

x = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, x)

B = np.array([[0, 1], [0, 0]])
P = np.array([[-1, 1], [2, 1]])
A = P@B@np.linalg.inv(P)

u = A[0, 0]*X + A[0, 1]*Y
v = A[1, 0]*X + A[1, 1]*Y

plt.quiver(X, Y, u, v)
```
